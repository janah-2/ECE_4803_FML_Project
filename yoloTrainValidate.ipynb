```python
from ultralytics import YOLO

# Load a model
model = YOLO('./trainingData/Run17/weights/best.pt')  # load a custom model

# Validate the model
metrics = model.val(data='trainRun.yaml',
                               imgsz=640,
                               batch=16,
                               conf=0.25,
                               iou=0.6,
                               device='0')
print(metrics)
print(metrics.box.map)    # map50-95
print(metrics.box.map50)  # map50
print(metrics.box.map75)  # map75
print(metrics.box.maps)   # a list contains map50-95 of each category
```
